**Count the missing values in each column. "True" respresents the missing values and "False" represnts the values exists in Dataset**

```python
for col_name in user_miss_val.columns.values.tolist(): # taking each column name into the existing variable
    print("Column Name", col_name)                     # within the loop we will be print each column name
    print(user_miss_val[col_name].value_counts())      # each column name to the method to get distinct value counts
    print('----------------------------------------------') # Line seperator
```

**User Info Data Set : Missing data findings** 
Based on the summary above, out of 4788 rows of data, and 2 columns containing missing data:

1. "PRODUCT_ADDONS": 2892 missing data
2. "USER_ROLE": 1858 missing data
