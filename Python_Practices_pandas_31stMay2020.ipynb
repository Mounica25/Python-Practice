```python
c_multiple = df1.groupby(['TX']).count()
c_multiple
```

```python
c_multiple = df1.groupby(['TX','VT']).sum()
c_multiple
```

```python
c_multiple.index # returns the multi index of a given dataframe
```

```python
c_multiple.groupby(['TX','VT']).size().reset_index().groupby('TX')[[0]].sum() 
```

```python
c_multiple.groupby(['TX','VT']).size().groupby(level=1).sum() # level indicates the column index
#We can groupby different levels of a hierarchical index using the level parameter:
```

```python
c_multiple['COUNTER'] = 1 
#For making a group of dataframe in pandas and counter,
#You need to provide one more column which counts the grouping, 
#let's call that column as, "COUNTER" in dataframe.
```

```python
c_multipleGrp = c_multiple.groupby(['TX','VT'])['COUNTER'].sum()
c_multipleGrp
```
